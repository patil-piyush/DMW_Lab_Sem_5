```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DMW/assignment-1/student.csv', header=1)
df.sample(10)
```

```python
unique_values = df['Pattern'].unique()
print(unique_values)
```

```python

df['Pattern'] = df['Pattern'].replace(['2020-21'], '2023')

display(df['Pattern'].value_counts())
```

```python
mode_value = df['Pattern'].mode()[0]

df['Pattern'] = df['Pattern'].fillna(mode_value)

df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
for column in df.columns:
    unique_values = df[column].unique()
    print(f'{column}-{unique_values}')
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for column in df.columns:
  df[column] = label_encoder.fit_transform(df[column])
```
